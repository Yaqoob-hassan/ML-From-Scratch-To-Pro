Importing Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Plotting with the help of numpy

```python
x = np.array([1, 2, 4, 8, 16])
y = np.array([0, 4, 5, 7,15])
plt.plot(x, y)
plt.show()
```

Understanding of Plotting

```python
fig = plt.figure()             # an empty figure with no Axes
fig, ax = plt.subplots()       # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one Axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

```python
#Plot is explain in the diagram. See in the image section.
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [5, 6, 7, 8])
plt.show()
```

Linestyling in plot

```python
x = np.array([1, 2, 4, 8, 16])
y = np.array([0, 4, 5, 7,15])
plt.plot(x, y, linestyle = 'dashed')
```

```python
x = np.array([1, 2, 4, 8, 16])
y = np.array([0, 4, 5, 7,15])
plt.plot(x, y, linestyle = 'dotted', color = 'black')
```

Giving Labels, setting font propertise and fontsize 

```python
x = np.array([1, 2, 4, 8, 16])
y = np.array([0, 4, 5, 7,15])
plt.plot(x, y, linestyle = 'dashed')
plt.suptitle("Plot", fontsize = 15, weight = 'bold') # The Title that show on the top 
plt.xlabel("X side", fontsize = 12) # X lable that show x data
plt.ylabel('Y side', fontsize = 14) #Y lable that show y data
plt.show()
```

Adding grid in the plot 

```python
x = np.array([1, 2, 4, 8, 16])
y = np.array([0, 4, 5, 7,15])
plt.plot(x, y, linestyle = 'dashed')
plt.suptitle("Plot", fontsize = 15, weight = 'bold') # The Title that show on the top 
plt.xlabel("X side", fontsize = 12) # X lable that show x data
plt.ylabel('Y side', fontsize = 14) #Y lable that show y data
plt.grid(color = 'green', linestyle = '--') # You can also specify which grid lines you want to show ( axis = 'x)
plt.show()
```

Advance version of the plots

```python
fig = plt.figure(figsize= (10, 5))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax1.hist(np.random.standard_normal(100), bins = 20, alpha = 0.3, color = 'yellow')
ax2.scatter(np.arange(30), np.arange(30)* 3 + np.random.standard_normal(30))
ax3.plot(np.random.standard_normal(100), linestyle = 'dashed', color = 'black')
```

Sample Graph workFlow

```python
x = np.linspace(0, 2, 200) # Sample data.

fig, ax = plt.subplots(figsize = (5, 2.7), label = 'constrained')
ax.plot(x, x, label = 'linear' ) # Plot some data on the Axes.
ax.plot(x, x**2 , label = 'quadratic') # Plot more data on the Axes...
ax.plot(x, x**3, label = 'cubic') # Plot more data on the Axes...
plt.title('Simple Plot') # Add title 
plt.xlabel('X label')# Add an x-label to the Axes.
plt.ylabel('Y label') # Add an y-label to the Axes.
plt.legend() # Add a legend.
plt.show()
```
